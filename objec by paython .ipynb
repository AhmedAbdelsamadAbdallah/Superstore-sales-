```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_excel('Sql New Data.xlsx')
```

```python
df
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
df['Shipping_Duration'] = (df['Ship_Date'] - df['Order_Date']).dt.days
```

```python
df['Month_Year'] = df['Order_Date'].dt.to_period('M')
```

```python
df
```

```python
Sales_time = df.groupby('Month_Year')['Sales'].sum()
```

```python
Sales_time
```

```python
import matplotlib.pyplot as plt
```

```python
plt.ylabel("Total Sales")
plt.xlabel("Month-Year")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
Category_sales = df.groupby(['Category', 'Sub_Category'])['Sales'].sum().sort_values()
```

```python
Category_sales.plot(kind='barh', figsize=(10, 8), title='Sales by Category and Sub-Category')
plt.xlabel("Sales")
plt.tight_layout()
plt.show()
```

```python
Segment_sales = df.groupby('Segment')['Sales'].sum()
```

```python
Segment_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title='Sales by Customer Segment')
plt.ylabel(' ')
plt.show()
```

```python
Top_products = df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
Top_products.plot(kind='barh', figsize=(10, 6), title='Top 10 Products by Sales')
plt.xlabel("Sales")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
Region_sales = df.groupby('Region')['Sales'].sum()
```

```python
Region_sales.plot(kind='bar', title='Sales by Region', color='teal')
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```
